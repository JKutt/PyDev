```python
import numpy as np
import matplotlib.pyplot as plt
from SimPEG import Utils
```

```python
X, Y = np.meshgrid(np.arange(-10,11,0.1), np.arange(-10,11,0.1))
Z = np.ones_like(X) * 10

x, y, z = X.reshape(-1), Y.reshape(-1), Z.reshape(-1)
```

```python
# Inducing field
H = Utils.matutils.spherical2cartesian(np.c_[50000, -np.deg2rad(90), np.deg2rad(0)])

# Define sphere at origin
R = 1 # meters
xc, yc, zc = 0, 0, 0
chi = 0.1

mx = np.ones_like(x) * H[0] * R**3 / 3 * chi
my = np.ones_like(y) * H[1] * R**3 / 3 * chi
mz = np.ones_like(z) * H[2] * R**3 / 3 * chi
M = np.c_[mx, my, mz]
print(X.shape)
dx = (x - xc)
dy = (y - yc)
dz = (z - zc)

rvec = np.c_[dx, dy, dz]
r = np.sqrt((dx)**2+(dy)**2+(dz)**2)
print(r.shape)

bx = (
    -mx/r**3 + 
    3 * np.sum(M*rvec, axis=1)/r**5*dx
)

by = (
    -my/r**3 + 
    3 * np.sum(M*rvec, axis=1)/r**5*dy
)

bz = (
    -mz/r**3 + 
    3 * np.sum(M*rvec, axis=1)/r**5*dz
)
```

```python
plt.figure(figsize=(12,8))
axs = plt.subplot(1,3,1)
plt.contourf(X, Y, bx.reshape(X.shape))
print(bx.reshape(X.shape).shape)
axs.set_aspect('equal')
axs.set_title("Bx")
axs = plt.subplot(1,3,2)
plt.contourf(X, Y, by.reshape(X.shape))
axs.set_aspect('equal')
axs.set_title("By")
axs = plt.subplot(1,3,3)
plt.contourf(X, Y, bz.reshape(X.shape))
axs.set_aspect('equal')
axs.set_title("Bz")

```

```python
bxx = (
    3*mx*dx/r**5 + 
    3*mx*dx/r**5 + 
    3 * np.sum(M*rvec, axis=1)/r**5 +
    -15 * np.sum(M*rvec, axis=1)/r**7*dx**2.
)
print(r.shape)
byy = (
    3*my*dy/r**5 + 
    3*my/r**5*dy + 
    3 * np.sum(M*rvec, axis=1)/r**5 +
    -15 * np.sum(M*rvec, axis=1)/r**7*dy**2.
)

bzz = (
    3*mz*dz/r**5 + 
    3*mz/r**5*dz + 
    3 * np.sum(M*rvec, axis=1)/r**5 +
    -15 * np.sum(M*rvec, axis=1)/r**7*dz**2.
)

bxy = (
    3*mx*dy/r**5 + 
    3*my*dx/r**5 + 
    -15 * np.sum(M*rvec, axis=1)/r**7*dx*dy
)

bxz = (
    3*mx*dz/r**5 + 
    3*mz*dx/r**5 + 
    -15 * np.sum(M*rvec, axis=1)/r**7*dx*dz
)

byz = (
    3*my*dz/r**5 + 
    3*mz*dy/r**5 + 
    -15 * np.sum(M*rvec, axis=1)/r**7*dz*dy
)
```

```python
plt.figure(figsize=(12,12))

axs = plt.subplot(3,3,1)
plt.contourf(X, Y, bxx.reshape(X.shape))
axs.set_aspect('equal')

axs = plt.subplot(3,3,2)
plt.contourf(X, Y, bxy.reshape(X.shape))
axs.set_aspect('equal')

axs = plt.subplot(3,3,3)
plt.contourf(X, Y, bxz.reshape(X.shape))
axs.set_aspect('equal')

axs = plt.subplot(3,3,5)
plt.contourf(X, Y, byy.reshape(X.shape))

axs = plt.subplot(3,3,6)
plt.contourf(X, Y, byz.reshape(X.shape))

axs = plt.subplot(3,3,9)
plt.contourf(X, Y, bzz.reshape(X.shape))

axs.set_aspect('equal')

```

```python
# calculating tensor components of q matrix
# xi = np.asarray([xc - R, xc + R])
# eta = np.asarray([yc - R, yc + R])
# zeta = np.asarray([zc - R, zc + R])
a = 0.1
R = 0.5
dx1 = x - (R)
dy1 = y - (R)
dz1 = z - (R)
dx2 = x + (R)
dy2 = y + (R)
dz2 = z + (R)
r1 = np.sqrt((dx1)**2+(dy1)**2+(dz1)**2)
r2 = np.sqrt((dx2)**2+(dy2)**2+(dz2)**2)
print(dx1[0], dx2[0], x[0], xc, R)
print(dy1[0], dy2[0], y[0], yc, R)
print(dz1[0], dz2[0], z[0], zc, R)
# r_ = np.asarray([dx, dy, dz])
# r = np.sqrt(np.sum(r_**2))

# print(r.shape)

# compactify argument calculations
arg1 = dy1 + dz1 + r1
arg2 = dx1 + dz1 + r1
arg3 = dx1 + r1
arg4 = dy1 + r1
arg5 = dz1 + r1
# print(arg1, arg2, arg3, arg4, arg5)

arg6 = dy2 + dz2 + r2
arg7 = dx2 + dz2 + r2
arg8 = dx2 + r2
arg9 = dy2 + r2
arg10 = dz2 + r2

# create m_x vector
qx_xx = 2 * (((dx2**2 - r2 * arg6) / (r2 * arg6**2 + dx2**2 * r2)) -
             ((dx1**2 - r1 * arg1) / (r1 * arg1**2 + dx1**2 * r1)))

qx_yy = dy2 / (r2 * arg10) - dy1 / (r1 * arg5)

qx_xy = 2 * (((dx2 * arg9) / (r2 * arg6**2 + (dx2**2) * r2)) -
             ((dx1 * arg4) / (r1 * arg1**2 + (dx1**2) * r1)))

qx_xz = 2 * (((dx2 * arg10) / (r2 * (arg7**2) + (dx2**2) * r2)) - 
             ((dx1 * arg5) / (r1 * (arg2**2) + (dx1**2) * r1)))

qx_yz = 1 / r2 - 1 / r1

# create m_y vector
qy_xx = (dx2 / (r2 * arg10)) - (dx1 / (r1 * arg5))

qy_yy = 2 * (((dy2**2 - r2 * arg7) / (r2 * arg7**2 + dy2**2 * r2))-
            ((dy1**2 - r1 * arg2) / (r1 * arg2**2 + dy1**2 * r1)))

qy_xy = dy2 / (r2 * arg10) - dy1 / (r1 * arg5)

qy_xz = 1 / r2 - 1 / r1

qy_yz = 2 * ((((dy2 * arg10) / (r2 * (arg7**2) + (dy2**2) * r2))) - 
             (((dy1 * arg5) / (r1 * (arg2**2) + (dy1**2) * r1))))

# create m_z vector
qz_xx = (dx2 / (r2 * arg9)) - (dx1 / (r1 * arg4))

qz_yy = dy2 / (r2 * arg8) - dy1 / (r1 * arg3)

qz_xy = 1 / r2 - 1 / r1

qz_xz = dz2 / (r2 * arg9) - dz1 / (r1 * arg4)

qz_yz = dz2 / (r2 * arg8) - dz1 / (r1 * arg3)

# qxx = qx_xx + qy_xx + qz_xx
qxx = np.dot(np.c_[qx_xx, qy_xx, qz_xx], H)
# qyy = qx_yy + qy_yy + qz_yy
qyy = np.dot(np.c_[qx_yy, qy_yy, qz_yy], H)
# qxy = qx_xy + qy_xy + qz_xy
qxy = np.dot(np.c_[qx_xy, qy_xy, qz_xy], H)
# qxz = qx_xz + qy_xz + qz_xz
qxz = np.dot(np.c_[qx_xz, qy_xz, qz_xz], H)
# qyz = qx_yz + qy_yz + qz_yz
qyz = np.dot(np.c_[qx_yz, qy_yz, qz_yz], H)
qzz = -qxx - qyy
```

```python
plt.figure(figsize=(12,12))

axs = plt.subplot(3,3,1)
plt.contourf(X, Y, qxx.reshape(X.shape))
axs.set_aspect('equal')

axs = plt.subplot(3,3,2)
plt.contourf(X, Y, qxy.reshape(X.shape))
axs.set_aspect('equal')

axs = plt.subplot(3,3,3)
plt.contourf(X, Y, qxz.reshape(X.shape))
axs.set_aspect('equal')

axs = plt.subplot(3,3,5)
plt.contourf(X, Y, qyy.reshape(X.shape))

axs = plt.subplot(3,3,6)
plt.contourf(X, Y, qyz.reshape(X.shape))

axs = plt.subplot(3,3,9)
plt.contourf(X, Y, qzz.reshape(X.shape))

axs.set_aspect('equal')
```
